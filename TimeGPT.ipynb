```python
import numpy as np
import pandas as pd
from nixtlats import TimeGPT

from tools import visualize
```

```python
timegpt = TimeGPT(token='')
```

```python
df = pd.read_csv("./data/air_passengers.csv")
train_df = df.iloc[:96, :]
test_df = df.iloc[96:, :]
pred_len = len(test_df)

print(train_df.shape, test_df.shape)
```

```python
pred_data = timegpt.forecast(df=train_df, h=48, freq='MS', time_col='timestamp', target_col='value')['TimeGPT'].values
visualize(train_df["value"].values, test_df["value"].values, pred_data)
```
