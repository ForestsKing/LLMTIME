```python
import numpy as np
import pandas as pd

from tools import visualize
from llmtime import Scaler, Serializer
```

```python
data = pd.read_csv("./data/air_passengers.csv")["value"].values
train_data = data[:96]
test_data = data[96:]

scaler = Scaler(train_data)
```

```python
with open('./result/ChatGPT_3_Space.txt','r')as f:
    pred_list = f.readlines()

serializer = Serializer(time_sep=',', bit_sep='', minus_sign='-',)

pred_data = []
for pred in pred_list:
    pred = pred.replace('\n', '')
    
    pred = serializer.deserialize(pred)[:48]
    pred = scaler.inv_transform(pred)
    pred_data.append(pred)

pred_data = np.mean(pred_data, axis=0)
visualize(train_data, test_data, pred_data)
```
